```python
from sklearn.model_selection import train_test_split
from app.service.machine_learning_functions.hyperparameter_search import (
    run_hyperparameter_search,
)
from app.service.machine_learning_functions.local_data_management import load_data
from app.service.machine_learning_functions.model_generation import evaluate_model

csv_path = "data/features.csv"
csv_path_kaggle = "data/kaggle.csv"
features, labels = load_data(csv_path, csv_path_kaggle)
features_train, features_validation, labels_train, labels_validation = train_test_split(
    features,
    labels,
    test_size=0.2,
    stratify=labels,
    random_state=42,
)
tuner = run_hyperparameter_search(
    features=features,
    labels=labels,
    max_trials=200,
    executions_per_trial=1,
    test_size=0.2,
    random_state=42,
    epochs=3,
    experiment_name="test_2",
)

best_models = tuner.get_best_models(num_models=5)
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=5)

for _i, best_model in enumerate(best_models):
    print(best_hyperparameters[_i].values)
    evaluate_model(best_model, features_validation, labels_validation)

best_hyperparameters = tuner.get_best_hyperparameters(num_trials=20)
hp_ranges = {}
for hp in best_hyperparameters:
    for name in hp.values.keys():
        values = hp.values[name]
        if name not in hp_ranges:
            hp_ranges[name] = {"min": values, "max": values}
        else:
            hp_ranges[name]["min"] = min(hp_ranges[name]["min"], values)
            hp_ranges[name]["max"] = max(hp_ranges[name]["max"], values)
print(hp_ranges)
```
